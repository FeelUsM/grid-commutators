```python
import math
def mean(arr):
    return round(1000*sum(arr)/len(arr))/1000
def disp(arr):
    return round(1000*math.sqrt(mean([x**2 for x in arr])-mean(arr)**2))/1000
```

```python
def nice_stat(file,Nthreads = 64):
    with open(file) as f:
        f.readline() # заголовок iostat
        f.readline() # заголовок iostat
        io_count = 0 # счетчик по строкам 'iostat:'
        top_count = 0
        tselect='5'  # выбранный поток для исследования

        genterms=[0]*Nthreads
        activeterms=[0]*Nthreads
        StageSort=[0]*Nthreads
        while True:
            line = f.readline()
            if not line:
                break
            if line.startswith('iostat:'):
                if io_count==1:
                    print('iostat:',line.split()[-1],'%'#)

                    #print(
                          ,('genterms',mean(genterms),min(genterms),max(genterms))
                    #print(
                          ,('activeterms',mean(activeterms),min(activeterms),max(activeterms))
                    #print(
                          ,('StageSort',mean(StageSort),min(StageSort),max(StageSort))
                         )
                    genterms=[0]*Nthreads
                    activeterms=[0]*Nthreads
                    StageSort=[0]*Nthreads
                io_count+=1
                if io_count==3: io_count = 0
            elif line.startswith('top:'):
                if line.startswith('top:top'):
                    top_count = 0
                    sline = line.split()
                    top_realtime= sline[2]

                if top_count>9:# or top_count==6 or top_count==7:
                    print(line[0:-1])

                if top_count==2:
                    sline = line.split()
                    top_us = sline[1]
                    top_id = sline[4]
                elif top_count==3:
                    sline = line.split()
                    top_free = sline[5]
                elif top_count==7:
                    sline = line[4:].split()
                    if len(sline)>10:
                        top_time= sline[10].split(':')
                        top_time= int(top_time[0])*60+float(top_time[1]) if len(top_time)==2 else ':'.join(top_time)
                        top_CPU = sline[8]
                        top_virt= sline[4]
                        top_res = sline[5]
                        top_shr = sline[6]
                    else:
                        top_time= 0
                        top_CPU = 0
                        top_virt= 0
                        top_res = 0
                        top_shr = 0
                    print('top:',top_realtime,top_time,(top_CPU,top_us,top_id),
                          (str(int(top_free[:-1])//1000000)+'g',top_virt,top_res,top_shr))
                top_count+=1

            elif line.startswith('StageSort'):
                sline = line.split()
                nt = int(sline[3])
                StageSort[nt-1]+=1
            else:
                sline = line.split()
                if len(sline)>2 and sline[2]=='iteration':
                    print('='*80)
                    print(line[0:-1])
                elif len(sline)>0 and (sline[0]=='.store' or sline[0]=='.sort') :
                    print(line[0:-1])
                    print('-'*80)
                elif len(sline)>2 and sline[0]=='Thread' and sline[2]=='reporting':
                    tn = int(sline[1])
                    if 0: # tn==tselect:
                        print(
                            'Thread',
                            ' '*(tn),
                            str(tn) if tn>9 else '0'+str(tn),
                            ' '*(Nthreads-tn),
                            'reporting'
                        )

                    line = f.readline()
                    if not line: break
                    sline1 = line.split()
                    line = f.readline()
                    if not line: break
                    sline2 = line.split()
                    line = f.readline()
                    if not line: break
                    sline3 = line.split()

                    A = False
                    if len(sline1)>5 and sline1[4]=='Generated' and sline1[5]=='terms':
                        genterms[tn-1]+=1
                    elif len(sline2)>2 and sline2[2]=='active':
                        activeterms[tn-1]+=1
                        A = True
                    else:
                        print(sline1)
                        print(sline2)
                        print(sline3)
                    if tn==0:
                        A = 'A' if A else '-'
                        print(' '.join(sline3),A)
                else:
                    print(repr(line))
```
