```python
def overlapping(nl,nr):
    lq = ['a'+str(i)+'?' for i in range(nl)];
    l  = ['a'+str(i)     for i in range(nl)];
    rq = ['b'+str(i)+'?' for i in range(nr)];
    r  = ['b'+str(i)     for i in range(nr)];
    if nl==0:   return 'id S()*S('+','.join(rq)+') = S('+','.join(r)+');'
    if nr==0:   return 'id S('+','.join(lq)+')*S() = S('+','.join(l)+');'
    if nl>=nr:
        rez=[]
        rez.append('0')
        for i in range(1,nr):
            rez.append('S('+','.join(r[:-i]+[l[j]+'*'+r[j-i] for j in range(i)]+l[i:])+')')
        rez.append('0')
        for i in range(nl-nr+1):
            rez.append('S('+','.join(l[:i]+[l[j+i]+'*'+r[j] for j in range(nr)]+l[nr+i:])+')')
        rez.append('0')
        for i in range(1,nr):
            rez.append('S('+','.join(l[:-i]+[l[j-i]+'*'+r[j] for j in range(i)]+r[i:])+')')
        return 'id S('+','.join(lq)+')*S('+','.join(rq)+') = '+'+'.join(rez)+';'
    else:
        rez=[]
        for i in range(1,nl):
            rez.append('S('+','.join(l[:-i]+[l[j-i]+'*'+r[j] for j in range(i)]+r[i:])+')')
        rez.append('0')
        for i in range(nr-nl+1):
            rez.append('S('+','.join(r[:i]+[l[j]+'*'+r[j+i] for j in range(nl)]+r[nl+i:])+')')
        rez.append('0')
        for i in range(1,nl):
            rez.append('S('+','.join(r[:-i]+[l[j]+'*'+r[j-i] for j in range(i)]+l[i:])+')')
        rez.append('0')
        return 'id S('+','.join(lq)+')*S('+','.join(rq)+') = '+'+'.join(rez)+';'
```

```python
f.close()
f = form.open("form -l",keep_log=100)
f.write('''
on statistics;
auto Symbols a,b;
Function S;
cFun den;
auto nTen m,n;
nten Id,c,c2,t,t2,ct,t2c2,c2t,t2c;
sym w,r23,k1,k2,k,l;
Fun Commute;
cfun P;  polyfun P;

Local A0 = k1*S(t) + k2*S(t2);
Local A1 = k1*S(c,c2) + k2*S(c2,c);
id k1 = -4/3*w*den(w);
id k2 = 4/3*den(w); * 1/(1-w)

id w^-1  = (-1-i_*r23)/2;
id w     = (-1+i_*r23)/2;
id den(w)= (3+i_*r23)/6;
id r23^2 = 3;

multiply P(1);
repeat id k?*P(l?)*S(?y) = P(k*l)*S(?y);

.sort

Local G0 = 0;
Local G1 = 1/4*Commute(A1,A0);


id Commute(a?,b?) = a*b - b*a;
'''+
''.join(overlapping(i,j) for i in range(3) for j in range(3))+
'''
argument;
'''+rules+
'''
endargument;
repeat id S(a?,?l) = a*S(?l);
repeat id m?*S(?l) = S(m,?l);
repeat id S(Id,?l) = S(?l);
repeat id S(?l,Id) = S(?l);

repeat id k?*P(l?)*S(?y) = P(k*l)*S(?y);
argument;  id r23^2 = 3; endargument;

.sort
''')
print('G1=',re.sub('([+-])S','\n\g<1>S',f.read('G1')))
```
